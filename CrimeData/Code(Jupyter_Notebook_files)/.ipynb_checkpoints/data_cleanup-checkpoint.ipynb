```python
import pandas as pd
```

```python
vc_pc_file_path="../Data_files/Crime_Data/CrimeData-2018.csv"
hate_crime_file_path="../Data_files/Crime_Data/hate_crime_data.csv"
unemployment_file_path="../Data_files/Unemployment_Data/unemployment_rate_bystate.csv"
# Load the csv file into a dataframe, Set the 4th row as header
vs_pc_raw_df=pd.read_csv(vc_pc_file_path,header=3) 
hate_crime_df=pd.read_csv(hate_crime_file_path, header=2)
unemployment_df=pd.read_csv(unemployment_file_path)
vs_pc_raw_df.head()
```

```python
# Rename the column name
column_renamed_df=vs_pc_raw_df.rename(columns={"Unnamed: 2":"Type of data"}) 
column_renamed_df.head()
```

```python
# Create a list of column names to drop
to_drop=["Unnamed: 13","Unnamed: 14"] 
# Drop the columns
reduced_df=column_renamed_df.drop(to_drop, axis=1) 
reduced_df.head()
```

```python
# Display the last 15 rows
reduced_df.tail(15)
```

```python
# Drop the last 10 rows
reduced_df=reduced_df.iloc[:-10] 
reduced_df.tail()
```

```python
# Replace NaN with the preceding values in State column
reduced_df["State"]=reduced_df["State"].fillna(method='ffill') 
reduced_df.head()
```

```python
# Create a 'Crime Rate per 100,000 inhabitants' dataframe
vc_per_100k_df=reduced_df.loc[reduced_df["Type of data"]=="Rate per 100,000 inhabitants",["State","Violent crime1","Property \ncrime"]].reset_index(drop=True)
```

```python
# Rename column names
vc_per_100k_df=vc_per_100k_df.rename(columns={"Violent crime1":"Violent Crime per 100,000","Property \ncrime":"Property Crime per 100,000"})

vc_per_100k_df.head()
```

```python
# Create a total crime dataframe for 50 states
total_vc_df=reduced_df.loc[reduced_df["Area"]=="State Total",["State","Violent crime1","Property \ncrime"]].reset_index(drop=True)

# Rename column names
total_vc_df=total_vc_df.rename(columns={"Violent crime1":"Total Violent Crime","Property \ncrime":"Total Property Crime"})

total_vc_df.head(5)
```

```python
# Create a population dataframe for 50 states
population_df=reduced_df.loc[reduced_df["Area"]=="State Total",["State","Population"]].reset_index(drop=True)

population_df.head()
```

```python
# Merge population and crime per 100k dataframe
combined_df=pd.merge(population_df,vc_per_100k_df,on="State")
combined_df.head()
```

```python
# Rename the incorrect state names
combined_df["State"]=combined_df.replace({"DISTRICT OF COLUMBIA3":"DISTRICT OF COLUMBIA","IOWA4":"IOWA","NORTH CAROLINA5":"NORTH CAROLINA"})
combined_df.head(10)
```

```python
# Loop through each row in the dataframe
for index,row in combined_df.iterrows():
    state=row["State"]
    # Split the state name, for each word uppercase the first letter, and lowercase the rest
    state_name = [word[0].upper() + word[1:].lower() for word in state.split()]
    s = " ".join(state_name)
    combined_df.loc[index,"State"]=str(s)

combined_df.head()
```

```python
hate_crime_df.head()
```

```python
# Select necessary columns
hate_crime_reduced_df=hate_crime_df.drop(['Number of\nparticipating\nagencies','Agencies\nsubmitting\nincident\nreports'], axis=1)
# Drop the last 4 unnecessary rows
hate_crime_reduced_df = hate_crime_reduced_df.drop([0,52,53,54,55])
# Rename the column
hate_crime_renamed_df=hate_crime_reduced_df.rename(columns={"Participating state/Federal":"State","Total\nnumber of\nincidents\nreported":"Hate Crime Reported","Population\ncovered":"Population covered in hate crime reporting"})
hate_crime_renamed_df["Hate Crime per 100,000"]=""
hate_crime_renamed_df.head()
```

```python
# Loop through each row in dataframe
for index,row in hate_crime_renamed_df.iterrows():
    # Remove ',' from the number, and convert it to integer
    total_hate_crime=int(row["Hate Crime Reported"].replace(",", ""))
    # Remove ',' from the number, and convert it to integer
    population_covered=int(row["Population covered in hate crime reporting"].replace(",", ""))
    # Calculate th hate crime per 100,000
    hate_crime_per_100k=round(total_hate_crime/(population_covered/100000),3)
    # Assign the values to new column
    hate_crime_renamed_df.loc[index,"Hate Crime per 100,000"]=hate_crime_per_100k


hate_crime_renamed_df.head()
```

```python
hate_crime_final_df=hate_crime_renamed_df[["State","Hate Crime per 100,000"]]
hate_crime_final_df.head()
```

```python
# Merge violent crime data and hate crime data
vc_hc_merged_df=pd.merge(combined_df,hate_crime_final_df, on="State")
vc_hc_merged_df.head()
```

```python
# Drop unnecessary rows and column
unemployment_reduced_df = unemployment_df.drop([0,1,2]).drop(["Rank"], axis=1)
# Rename column
unemployment_final_df = unemployment_reduced_df.rename(columns={"2018":"Unemployment Rate"})
unemployment_final_df.head()
```

```python
# Merge the unemployment data and crime data
final_merged_df = pd.merge(vc_hc_merged_df,unemployment_final_df, on="State")
final_merged_df.head(10)
```

```python
final_merged_df.to_csv("../Data_files/Final_Data/merged_data.csv", index=False)
```
