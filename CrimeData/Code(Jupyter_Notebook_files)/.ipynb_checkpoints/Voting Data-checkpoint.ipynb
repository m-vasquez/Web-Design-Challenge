```python
import pandas as pd
import plotly.express as px
```

```python
voting_data = "Voting Data- potential electoral datasets/states_by_color.csv"
```

```python
voting_data_df = pd.read_csv(voting_data)
voting_data_df
```

```python
states_df = voting_data_df.sort_values("State")
states_abbrev = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']
states_df['State Abrreviation'] = states_abbrev
states_df
```

```python
red_states_df = states_df[states_df["pvi"] > 3]
red_states_df
```

```python
blue_states_df = states_df[states_df["pvi"] < -1]
blue_states_df
```

```python
purple_states_df = states_df[(states_df["pvi"] >= -1) & (states_df["pvi"] <= 3)]
purple_states_df
```
