```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
file_path="../../Data_files/Final_Data/Crime_Data_Analysis(combined_info).csv"
data=pd.read_csv(file_path)
data.head()
```

```python
data.dtypes
```

```python
def lin_regress(x,y):
    (slope,intercept,rvalue,pvalue,stderr)=stats.linregress(x,y) # Calculate the linear regression
    regress=slope*x+intercept # Calculate the new y values using the linear equation
    lin_equation=f"y={round(slope,3)}*x+{round(intercept,2)}"
    correlation=stats.pearsonr(x,y)
    
    print(f"The correlation is {round(correlation[0],3)}")
    print(f"The R squared is {round(rvalue**2,3)}")
    print(f"The p-value is {round(pvalue,3)}")
    #Display the line and scatter plot
    plt.figure(figsize=(10,6))
    plt.plot(x,regress,color="red")
    plt.annotate(lin_equation,xy_coord,fontsize="xx-large",color="red",fontweight="semibold")
    plt.scatter(x,y,c="blue", alpha=0.4, edgecolor="black") 
    plt.annotate(f"Correlation: {round(correlation[0],3)}\nR-squared: {round(rvalue**2,3)}\nP-value: {round(pvalue,3)}",\
                coord,fontsize="large",color="black")
```

```python
xy_coord=(3000,600)
coord=(1200,750)
lin_regress(data["Property Crime per 100,000 inhabitants"],data["Violent Crime per 100,000 inhabitants"])
plt.xlabel("Property Crime per 100,000")
plt.ylabel("Violent Crime per 100,000")
```

```python
x=data["Hate Crime per 100,000 inhabitants"]
y=data["Violent Crime per 100,000 inhabitants"]

xy_coord=(5,600)
coord=(4,750)
#lin_regress(data["Hate Crime per 100,000 inhabitants"],data["Violent Crime per 100,000 inhabitants"])

(slope,intercept,rvalue,pvalue,stderr)=stats.linregress(x,y) # Calculate the linear regression
regress=slope*x+intercept # Calculate the new y values using the linear equation
lin_equation=f"y={round(slope,3)}*x+{round(intercept,2)}"
correlation=stats.pearsonr(x,y)
    
print(f"The correlation is {round(correlation[0],3)}")
print(f"The R squared is {round(rvalue**2,3)}")
print(f"The p-value is {round(pvalue,3)}")
#Display the line and scatter plot
plt.figure(figsize=(10,6))
plt.plot(x,regress,color="red")
plt.annotate(lin_equation,xy_coord,fontsize="xx-large",color="red",fontweight="semibold")
plt.annotate(f"Correlation: {round(correlation[0],3)}\nR-squared: {round(rvalue**2,3)}\nP-value: {round(pvalue,3)}",\
            coord,fontsize="large",color="black")
plt.scatter(x,y,c="blue", alpha=0.4, edgecolor="black",facecolor=['red','blue','purple']) 
plt.title("Hate Crime vs Violent Crime per 100,000 inhabitants")
plt.xlabel("Hate Crime per 100,000")
plt.ylabel("Violent Crime per 100,000")
#plt.savefig("../../example_figure/Scatter:hate_vs_violent_per_100k.png")
```

```python
red_violent_crime =data.loc[data["Affiliated Party Color"]=="Red","Violent Crime per 100,000 inhabitants"]
red_hate_crime=data.loc[data["Affiliated Party Color"]=="Red","Hate Crime per 100,000 inhabitants"]

purple_violent_crime=data.loc[data["Affiliated Party Color"]=="Purple","Violent Crime per 100,000 inhabitants"]
purple_hate_crime=data.loc[data["Affiliated Party Color"]=="Purple","Hate Crime per 100,000 inhabitants"]

blue_violent_crime= data.loc[data["Affiliated Party Color"]=="Blue","Violent Crime per 100,000 inhabitants"]
blue_hate_crime= data.loc[data["Affiliated Party Color"]=="Blue","Hate Crime per 100,000 inhabitants"]

#Assign x and y for regression
x=data["Hate Crime per 100,000 inhabitants"]
y=data["Violent Crime per 100,000 inhabitants"]

xy_coord=(5,600)
coord=(4,750)

(slope,intercept,rvalue,pvalue,stderr)=stats.linregress(x,y) # Calculate the linear regression
regress=slope*x+intercept # Calculate the new y values using the linear equation
lin_equation=f"y={round(slope,3)}*x+{round(intercept,2)}"
correlation=stats.pearsonr(x,y)

#Display the line plot
plt.figure(figsize=(10,6))
plt.plot(x,regress,color="red")
plt.annotate(lin_equation,xy_coord,fontsize="xx-large",color="red",fontweight="semibold")
plt.annotate(f"Correlation: {round(correlation[0],3)}\nR-squared: {round(rvalue**2,3)}\nP-value: {round(pvalue,3)}",\
            coord,fontsize="large",color="black")

#Display the scatter plots with different colors
plt.scatter(x=red_hate_crime,y=red_violent_crime, c="red", label="Red States")
plt.scatter(x=blue_hate_crime,y=blue_violent_crime, c="blue", label="Blue States")
plt.scatter(x=purple_hate_crime,y=purple_violent_crime, c="purple", label="Purple States")
plt.legend()

plt.title("Hate Crime vs Violent Crime per 100,000 inhabitants")
plt.xlabel("Hate Crime per 100,000 inhabitants")
plt.ylabel("Violent Crime per 100,000 inhabitants")
plt.savefig("../../example_figures/Scatter:hate_vs_violent_per_100k.png")
plt.show()
```
